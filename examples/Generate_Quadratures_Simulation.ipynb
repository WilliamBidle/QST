# Declare Parameters

```python
num_data_points = 1000 # the number of data points to simulate
save_data = True # whether or not to save the data
filename = 'Test_sim' + str(num_data_points) # save the data to 'filename'

x = np.linspace(-10, 10, num_data_points) # the domain to consider 
theta = np.linspace(0, 2*np.pi, num_data_points) # the phase space scanned by the piezo
```
